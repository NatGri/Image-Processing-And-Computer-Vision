```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('../Practice-Images/goldfish.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

print(img.shape)
plt.imshow(img, cmap='gray'), plt.title('Original')
plt.show()

dimension = (400,400)

img_resized = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)

print(img_resized.shape)

plt.imshow(img_resized, cmap='gray'), plt.title('Resized')
plt.show()
```

```python
import imutils

imutils_rotate = imutils.rotate(img_resized, -30)
plt.imshow(imutils_rotate, cmap='gray'), plt.title('imutils.rotate')
plt.show()

imutils_rotate_bound = imutils.rotate_bound(img_resized, -30)
plt.imshow(imutils_rotate_bound, cmap='gray'), plt.title('imutils.rotate')
plt.show()
```

```python
# Task 2:
added_img = cv2.addWeighted(imutils_rotate, 0.2, img_resized, 0.4, 0)
plt.imshow(added_img, cmap='gray'), plt.title('addWeighted')
plt.show()
```

```python
while True:
    print('Please Enter a Value for image A: 0-100\n')
    t_value = input()
    t_value = int(t_value)
    t_value = t_value/100
    try:
        added_img = cv2.addWeighted(imutils_rotate, t_value, img_resized, 1-t_value, 0)
        plt.imshow(added_img, cmap='gray'), plt.title('addWeighted')
        plt.show()
    except:
        print('Please enter only numerical values!')
        continue
    if t_value < 0 or t_value > 100:
        print('Please keep the number between 0-100!')
        continue
    break
```
