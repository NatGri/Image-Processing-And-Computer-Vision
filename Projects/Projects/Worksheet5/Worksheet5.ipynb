Write code to read in a colour RGB/BGR image and convert it to grayscale, using 3 different methods:
a)	using opencvâ€™s cv2.COLOR_BGR2GRAY parameter
b)	convert to HSV/HSB and extract one of the channels (which?)
c)	using the formula below, for each pixel:
`GS_value = 0.298936*R + 0.587043*G + 0.114021*B`
Display the 3 results you get and compare the images.


```python
import cv2
import matplotlib.pyplot as plt

# Import and Convert to RGB
image = cv2.imread('../Practice-Images/lena.jpg')
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# A:
cv2Gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# B:
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
hsv = hsv[:,:,2]

# C:
# image is in BGR so Red is index 2 in the array [B,G,R]
GS_value = 0.298936*image[:,:,2] + 0.587043*image[:,:,1] + 0.114021*image[:,:,0]

# Show:
plt.axis('off')
plt.imshow(cv2Gray, cmap="gray", vmin=0, vmax=255)
plt.show()
plt.axis('off')
plt.imshow(hsv, cmap="gray")
plt.show()
plt.axis('off')
plt.imshow(GS_value, cmap="gray")
plt.show()
```
