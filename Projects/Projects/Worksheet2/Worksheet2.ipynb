### Colour Splitting

Write a program which `reads/loads` an RGB Image and displays the red,green and blue channels separately, each one of them as an RGB Image.

```python
import matplotlib.pyplot as plt
import cv2

CS = cv2.cvtColor(cv2.imread('../Practice-Images/lena.jpg'),cv2.COLOR_BGR2RGB)

CS_R = CS.copy()
CS_R[:,:,(1,2)] = 0

CS_G = CS.copy()
CS_G[:,:,(0,2)] = 0

CS_B = CS.copy()
CS_B[:,:,(0,1)] = 0

###############
plt.imshow(CS)
plt.show()

plt.imshow(CS_R)
plt.show()

plt.imshow(CS_G)
plt.show()

plt.imshow(CS_B)
plt.show()
```

### Image Brightening

Write a program which loads an RGB Image and brightens each pixel by a value of `5 (current brightness + 5)`. Do this by applying an addition operation between a matrix and a scalar.

```python
IB = cv2.cvtColor(cv2.imread('../Practice-Images/lena.jpg'),cv2.COLOR_BGR2RGB)

IB_B = IB.copy()
# Scalar Operation
IB_B += 5

###############
plt.imshow(IB)
plt.show()

plt.imshow(IB_B)
plt.show()
```

### Image Trimming

Write a program which loads an RGB image and trims out a margin of 20 pixels from all around the image. Do this by using the colon notation. Display the new image.

### Image Flipping

Write a program which loads an RGB image and flips it both vertically and horizontally. Do this by using the colon notation. Display the new Image.

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

IF = np.array(Image.open('../Practice-Images/lena.jpg'))
IF_UD = np.flipud(IF)
IF_LR = np.fliplr(IF)
IF_C = np.fliplr(IF_UD)

###############
plt.imshow(IF)
plt.show()

plt.imshow(IF_UD)
plt.show()

plt.imshow(IF_LR)
plt.show()

plt.imshow(IF_C)
plt.show()
```
