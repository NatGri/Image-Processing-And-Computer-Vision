```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Task 1:
img_t = cv2.imread("./trees.png", 0)
img_t = cv2.cvtColor(img_t, cv2.COLOR_BGR2RGB)

ret, thresh1 = cv2.threshold(img_t, 255*0.25, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img_t, 255*0.5, 255, cv2.THRESH_BINARY)
ret, thresh3 = cv2.threshold(img_t, 255*0.75, 255, cv2.THRESH_BINARY)


titles = ['Original image', 'Binary 25%', 'Binary 50%', 'Binary 75%']

images = [img_t, thresh1, thresh2, thresh3]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
img_c = cv2.imread("./creative-proposal.jpg", 0)

ret, thresh1 = cv2.threshold(img_c, 255*0.5, 255, cv2.THRESH_BINARY)
thresh2 = cv2.adaptiveThreshold(img_c, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
thresh3 = cv2.adaptiveThreshold(img_c, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

titles = ['Original image', 'Binary 50%', 'Binary Adaptive', 'Adaptive Binary Gaussian']

images = [img_c, thresh1, thresh2, thresh3]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

## Task 3:

```python
img_j = cv2.imread("./j.png", 0)
img_j = cv2.cvtColor(img_j, cv2.COLOR_BGR2RGB)

ret, thresh1 = cv2.threshold(img_j, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)

eroded = cv2.erode(thresh1, kernel, iterations = 1)
dilated = cv2.dilate(thresh1, kernel, iterations = 1)

titles = ['Original image', 'Binary 50%', 'Eroded', 'Dilated']
images = [img_j, thresh1, eroded, dilated]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(1,4,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

## Task 4 - Opening and Closing

```python
opening = cv2.imread("./opening_original.png")
closing = cv2.imread("./closing_original.png")

ret, opening1 = cv2.threshold(opening, 127, 255, cv2.THRESH_BINARY)
ret, closing1 = cv2.threshold(closing, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)

opening2 = cv2.morphologyEx(opening1, cv2.MORPH_OPEN, kernel)
closing2 = cv2.morphologyEx(closing1, cv2.MORPH_CLOSE, kernel)

titles = ['Original Opening', 'Opening Morphed', 'Original Closing', 'Closing Morphed']
images = [opening1, opening2, closing1, closing2]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
