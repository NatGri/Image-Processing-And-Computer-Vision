## Task 3:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_j = cv2.imread("./j.png", 0)
img_j = cv2.cvtColor(img_j, cv2.COLOR_BGR2RGB)

ret, thresh1 = cv2.threshold(img_j, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)

eroded = cv2.erode(thresh1, kernel, iterations = 1)
dilated = cv2.dilate(thresh1, kernel, iterations = 1)

titles = ['Original image', 'Binary 50%', 'Eroded', 'Dilated']
images = [img_j, thresh1, eroded, dilated]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(1,4,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

## Task 4 - Opening and Closing

```python
opening = cv2.imread("./opening_original.png")
closing = cv2.imread("./closing_original.png")

ret, opening1 = cv2.threshold(opening, 127, 255, cv2.THRESH_BINARY)
ret, closing1 = cv2.threshold(closing, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)

opening2 = cv2.morphologyEx(opening1, cv2.MORPH_OPEN, kernel)
closing2 = cv2.morphologyEx(closing1, cv2.MORPH_CLOSE, kernel)

titles = ['Original Opening', 'Opening Morphed', 'Original Closing', 'Closing Morphed']
images = [opening1, opening2, closing1, closing2]

plt.figure(figsize=(16,12))

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
